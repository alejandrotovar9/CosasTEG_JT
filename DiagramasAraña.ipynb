```python
import plotly.graph_objects as go
from google.colab import files

categories=['Disponibilidad','Precio','GPIO',
           'Conectividad', 'Memoria', 'Velocidad de reloj']

colors = ['black']

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=[10, 9, 5, 9, 6, 5, 10],  # Add an additional value to connect the last point to the first point
      theta=categories + [categories[0]],  # Add the first category value at the end
      fill='none',
      mode='lines',
      name='Raspberry Pi Pico W'
))
fig.add_trace(go.Scatterpolar(
      r=[9, 9.5, 8, 9, 8, 8, 9],
      theta=categories + [categories[0]],
      fill='toself',
      #mode='lines',
      name='Espressif ESP32'
))
fig.add_trace(go.Scatterpolar(
      r=[5, 6, 8, 3, 10, 10, 5],
      theta=categories + [categories[0]],
      fill='none',
      mode='lines',
      name='Teensy 4.0'
))
fig.add_trace(go.Scatterpolar(
      r=[6, 7, 10, 4, 7, 7.5, 6],
      theta=categories + [categories[0]],
      fill='none',
      mode='lines',
      name='STM32F4 Discovery'
))
fig.add_trace(go.Scatterpolar(
      r=[7, 7, 9.5, 4, 6, 8, 7],
      theta=categories + [categories[0]],
      fill='none',
      mode='lines',
      name='STM32L432 Nucleo'
))
fig.add_trace(go.Scatterpolar(
      r=[8, 8, 8, 8, 8, 8, 8],
      theta=categories + [categories[0]],
      fill='none',
      #fill='toself'
      mode='lines',
      line=dict(color='black', width=2.5),
      name='Ideal'
))

# Update layout with title

fig.update_layout(
  paper_bgcolor='white',
  title={
        'text': "Diagrama de araña para selección de tarjeta de desarrollo basada en MCU",
        'x': 0.5,
        'y': 0.9,
        'xanchor': 'center',
        'yanchor': 'middle'
    },
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[10, 3]
    )),
  showlegend=True,
  legend=dict(
    orientation="h",
    yanchor="top",
    y=-0.2,
    xanchor="center",
    x=0.5
  )
)

fig.show()
```
